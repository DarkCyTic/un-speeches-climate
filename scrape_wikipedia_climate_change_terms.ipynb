```python
import requests

response = requests.get(
	url="https://en.wikipedia.org/wiki/Glossary_of_climate_change",
)
print(response.status_code)
```

```python
import re
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')

terms = []

for dl in soup.find_all('dl', {"class": "glossary"}):
    for dt in dl.find_all('dt', {'class': 'glossary'}):
        found = dt.getText()
        found = found.replace(')', '(')
        found = found.split('(')
        
        for word in found:
            if word:
                terms.append(word)
```
